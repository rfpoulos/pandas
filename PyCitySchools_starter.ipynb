```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
total_schools = school_data["school_name"].count()
total_students = student_data["student_name"].count()
total_budget = school_data["budget"].sum()
average_math = student_data["math_score"].mean()
average_reading = student_data["reading_score"].mean()
pass_score = (average_math + average_reading) / 2

pass_math_condition = student_data["math_score"] > 70
pass_math_count = student_data[pass_math_condition]["student_name"].count()
percent_pass_math = pass_math_count / total_students

pass_reading_condition = student_data["reading_score"] > 70
pass_reading_count = student_data[pass_reading_condition]["student_name"].count()
percent_pass_reading = pass_reading_count / total_students

data = {
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math": [average_math],
    "Average Reading": [average_reading],
    "Pass Score": [pass_score],
    "Percent Students Pass Math": [percent_pass_math],
    "Percent Students Pass Reading": [percent_pass_reading]
}
print(data)
summary_df = pd.DataFrame(data)

summary_df.head()


```

## Top Performing Schools (By Passing Rate)


* Sort and display the top five schools in overall passing rate

```python
grouped = school_data_complete.groupby(["school_name"])
average_math = grouped["math_score"].mean()
total_students = grouped["student_name"].count()
budgets = grouped["budget"].unique()
size = grouped["size"].unique()
per_student_budgets = budgets / total_students
average_reading = grouped["reading_score"].mean()

pass_math = school_data_complete[school_data_complete["math_score"] >= 70]
pass_math_grouped = pass_math.groupby(["school_name"])["student_name"].count() / total_students

pass_reading = school_data_complete[school_data_complete["reading_score"] >= 70]
pass_reading_grouped = pass_reading.groupby(["school_name"])["student_name"].count() / total_students

data = {
    "Total Students": total_students,
    "Budget": budgets,
    "Per Student Budget": per_student_budgets,
    "Average Math": average_math,
    "Average Reading": average_reading,
    "Percent Pass Math": pass_math_grouped,
    "Percent Pass Reading": pass_reading_grouped,
    "Pass Average": (pass_math_grouped + pass_reading_grouped) / 2,
    "School Size": size
}

school_summary = pd.DataFrame(data)
school_summary.head()
```
